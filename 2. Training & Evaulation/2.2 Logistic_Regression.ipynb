```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pickle
df = pd.read_csv('/workspaces/2025SE_Gianfranco.M_CSAI/1. Data/csgo_round_snapshots.csv')
```

import testing data VIA PICKLE !

```python
with open('split_data.pkl', 'rb') as f:
    X_train, X_test, y_train, y_test = pickle.load(f)
```

feature engineer for CT and T money difference

```python
train_data = pd.DataFrame(X_train[:, [10, 11]], columns=['ct_money', 't_money'])
train_data['round_winner'] = y_train
train_data['money_difference'] = train_data['ct_money'] - train_data['t_money']
```

```python
plt.figure(figsize=(10, 10))
sns.boxplot(x='round_winner', y='money_difference', data=train_data)
plt.xlabel('Round Winner')
plt.ylabel('money difference')
plt.title('Round Winner vs. money_diffence')
plt.show()
```
